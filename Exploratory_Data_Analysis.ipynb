**QUESTION 5**
-  Identifying and describing any significant outliers in the data and Data Visualization

Identifying outliers is crucial in data analysis because outliers can significantly affect the results of statistical analyses, potentially leading to misleading conclusions.



**Step 1**- Visualizing the Data to Identify Outliers

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Create box plots for each numerical feature to identigy outliers
plt.figure(figsize=(15, 10))
sns.boxplot(data=data[['Temp_C', 'Dew Point Temp_C', 'Rel Hum_%', 'Wind Speed_km/h', 'Visibility_km', 'Press_kPa']])
plt.title('Box Plots of Numerical Features')
plt.show()
```

**Explanation**: **BOX PLOTS**
-  The box plot shows the distribution of the data and highlights the outliers. The "whiskers" in a box plot extend to the minimum and maximum values within 1.5 times the interquartile range (IQR). Any points beyond these whiskers are considered outliers.

**Analysis Explanation**
-  "In this analysis, I identified and examined significant outliers in the dataset's numerical features. Outliers were detected using both box plots and the interquartile range (IQR) method, which revealed several data points that fall outside the typical range of values.

-  For instance, the temperature (Temp_C) feature showed outliers on both ends of the distribution, with extremely low temperatures (as low as -23.3°C) and high temperatures (up to 33°C) being identified as outliers. These extreme values could represent unusual weather conditions, such as severe cold spells or heatwaves. Similarly, wind speed (Wind Speed_km/h) exhibited outliers at the high end, with wind speeds reaching up to 83 km/h, which could indicate strong wind events or storms.

-  In the case of visibility (Visibility_km), outliers were observed on the lower end, with some data points showing visibility as low as 0.2 km, possibly due to foggy or stormy conditions. Atmospheric pressure (Press_kPa) had relatively fewer outliers, with some slightly high or low values that might correspond to unusual weather patterns.

-  These outliers are significant as they represent extreme weather conditions that are not typical of the general dataset. Identifying these outliers is crucial because they could influence the analysis, particularly in statistical modeling, where outliers might skew the results or suggest the need for further investigation into these extreme conditions."

**QUESTION 6**
-  Plot time series graphs to visualize trends over time. Highlight any notable patterns or seasonal variations.

```python
data = pd.read_csv("1. Weather Data.csv")
```

```python
print(data.head())
```

```python
print(data['Date/Time'].dtype)
```

**Plotting the Graphs**

```python
# Temperature over Time Plot
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Temp_C'], label='Temperature (°C)', color='orange')
plt.title('Temperature Over Time')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Relative Humidity over time Plot
plt.figure(figsize=(14, 7))
plt.plot(data.index, data['Rel Hum_%'], label='Relative Humidity (%)', color='blue')
plt.title('Relative Humidity Over Time')
plt.xlabel('Date')
plt.ylabel('Relative Humidity (%)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Wind Speed over Time Plot
plt.figure(figsize=(14, 7))
plt.plot(data.index, data['Wind Speed_km/h'], label='Wind Speed (km/h)', color='green')
plt.title('Wind Speed Over Time')
plt.xlabel('Date')
plt.ylabel('Wind Speed (km/h)')
plt.grid(True)
plt.legend()
plt.show()
```

** Visualization Overview**
-  In this analysis, I plotted time series graphs for key weather variables, including temperature, relative humidity, and wind speed, to visualize trends over time.

-  The temperature (Temp_C) graph reveals clear seasonal variations, with temperatures rising during the warmer months and dropping during the colder months. This cyclic pattern is typical of seasonal changes and aligns with expectations for weather patterns in most regions. The peaks and troughs correspond to summer and winter, respectively.

-  The relative humidity (Rel Hum_%) graph shows a less pronounced but noticeable seasonal trend, with humidity levels tending to be higher during the colder months and lower during the warmer months. This pattern could be due to the interaction between temperature and humidity, where cold air tends to hold less moisture, leading to higher relative humidity.

-  The wind speed (Wind Speed_km/h) graph displays fluctuations throughout the year, with occasional spikes that may indicate stormy weather or high-wind events. While there isn't a strong seasonal pattern in wind speed, the spikes are notable and could be investigated further to understand their causes.

-  Overall, these time series plots reveal important trends and patterns in the data, with temperature showing strong seasonal variations and other features like relative humidity and wind speed displaying more subtle patterns. Identifying these trends is crucial for understanding the underlying dynamics of the weather data and can inform further analysis or forecasting efforts.

**QUESTION 7**
- Creating Correlation matrices and heatmaps to identify relationships between weather parameters.

```python
# Convert 'Date/Time' to datetime objects if it's not already
data['Date/Time'] = pd.to_datetime(data['Date/Time'])

# Extract numeric features for correlation calculation
numeric_data = data.select_dtypes(include=['number'])

# Calculate the correlation matrix
correlation_matrix = numeric_data.corr()

# Display the correlation matrix
print(correlation_matrix)
```

-  Heatmap to Visualize the Correlation Matrix

```python
# Size of the plot
plt.figure(figsize=(10, 8))

# Creating a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# Adding a title to the heatmap
plt.title('Correlation Heatmap of Weather Parameters')

# Plot
plt.show()
```
